## 1.1 What is Feature Extraction in NLP (Signal vs. Noise)

* Isolating Useful Information: the aim is to isolate the useful or relevant information from a mass of data. In NLP, feature extraction is about selecting only those aspects of the data that are useful for a particular task (e.g., sentiment analysis, machine translation, etc.).

* Improving Accuracy: Removing irrelevant features in NLP usually leads to better results, more accurate machine learning models.

* Optimization: Iterative process, requiring several rounds of refinement to optimize performance.

* Sometimes noise is signal and sometimes noise is actually signal in natural language processing. We will see this in module 2 when looking at capitalization as a useful feature as compared to times during text preprocessing when it is often more noise than signal.
